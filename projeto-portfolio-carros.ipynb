```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Lendo o DataSet

```python
df = pd.read_csv("/kaggle/input/new-york-cars-big-data-2023/Car_Rates.csv")
df
```

# Renomeando as Colunas

```python
df.rename(columns = {'Car_name':'Nome_Carro', 'Num_of_reviews': 'Numero_Votos', 'General_rate': 'Notas_Geral', 'Comfort': 'Conforto',
       'Interior design': 'Nota_Desing_Interior', 'Performance': 'Performace', 'Value for the money':'Nota_Preco_Justo',
       'Exterior styling': 'Nota_Desing_Exterior', 'Reliability': 'Confiabilidade', 'Year':'Ano', 'Brand':'Marca', 'Model':'Modelo'}, inplace = True)
df
```

# Removendo as Linhas onde o Numero_Votos Ã© NaN

```python
df.shape
```

```python
df = df[~df['Numero_Votos'].isna()]
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

# Ã‰ possÃ­vel observar que o nÃºmero de votos Ã© muito baixo, e a Confiabilidade nÃ£o foi informada, entÃ£o vamos excluÃ­-la

```python
df[df['Confiabilidade'].isna()]
```

```python
df = df[~df['Confiabilidade'].isna()]
df
```

# Agora vamos analisar os dados, e ver se existem dados que possam ser possÃ­veis outliers

```python
df.describe()
```

# Vamos analisar se existem muios carros com com pouco Numero_Votos

# Vamos tambÃ©m excluir essas linhas com Numero_Votos <= 2

```python
df = df.drop(index = df.query("Numero_Votos <= 2").index)
```
